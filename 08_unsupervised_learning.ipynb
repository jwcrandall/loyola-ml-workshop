# Unsupervised Learning
## Background
Unsupervised learning is a type of machine learning method which involves the process of learning patterns from the input data where the **ground truth** is either not known or not referred to.

The most common unsupervised learning method is the cluster analysis, which is used for **exploratory** data analysis to find grouping(s) or hidden patterns in data. This grouping is also called a **partition** in Computer Science. The clusters are identified using a measure of **similarity** which is defined upon metrics such as Euclidean or probabilistic distance. The cluster analysis groups the data that has not been labeled, not classified or not categorized. Instead of responding to feedback (target/dependent variable), cluster analysis identifies commonalities in the data and reacts based on the presence or absence of such commonalities in each new piece of data.

The simplest form of clustering can be considered as the **binning** process as in data preprocessing, where the input numerical values are grouped according to the ranges of the bins.
* Supervised learning intends to infer a conditional probability distribution $P_X(x|y)$ conditioned on the labels $y$ of input data
* Unsupervised learning intends to infer an a priori probability distribution $P_X(x)$

Unsupervised learning is the first form of learning which is much more common in the brain.

Clustering is the main method of unsupervised learning which can answer questions like,
* What is the "normal" in the dataset? What is the baseline?
* Which data points are abnormal, anomalous in the dataset, if there are any?
* Is there a probability distribution that can model the input dataset?
* How many clusters are there in the dataset naturally?
* Anomaly detection

Types of clustering:  
**Hard clustering:** Each data point can be assigned to one cluster only  
**Soft clustering:** Also known as **fuzzy clustering**, where each point can have a **membership** value to any cluster.

## Major Clustering Algorithm 
### Distance/Similarity Measures
All clustering algorithms employ some form of similarity or a distance measure.

A distance metric, $d: X \times X \rightarrow[0, \infty)$, satisfies the following:

* non-negative $(x, y) \geq 0$                       
* identity $d(x, y)=0 \Longleftrightarrow x=y$
* symmetry $d(x, y)=d(y, x)$
* triangle inequality $d(x, y) \leq d(x, z)+d(z, y)$

Define $d_{ik} = $ the distance between $x_i$ and the cluster center $c_k$ where $k = 1, ..., K$, and $x \in \mathbb{R}^{N \times M}$ and $c_{k} \in \mathbb{R}^{M}$

* Euclidean distance $d_{i k}=\sqrt{\sum_{j=0}^{M}\left(x_{i j}-c_{k j}\right)^{2}}$ â€ƒ
* Manhattan distance $d_{i k}=\sum_{j=0}^{M}\left|x_{i j}-c_{k j}\right|$
* Chebyshev distance $d_{i k}=\max_{j}\left|x_{i j}-c_{k j}\right|$
* Cosine similarity $d_{i k}=\cos (\theta)=\frac{x_{i} \cdot c_{k}}{\left\|x_{i}\right\|\left\|c_{k}\right\|}$ 
* Jaccard index $d_{i k}=\frac{x_{i} \cap c_{k}}{x_{i} \cup c_{k}}$

### K-means Clustering Algorithm
K-means clustering aims to partition $N$ data points into $K$ number of clusters in which each observation belongs to the cluster with the nearest mean (i.e. cluster centroid).
1. Randomly pick $K$ data points from the input data set as initial cluster centers, call $\mu_k$ where $k=1,..,K$ 
2. Assign each data point $x_i$ to the cluster via the nearest centroid $\mu_k$ where $k=1,..,K$.
3. Recompute (move) the centroids to the center of the data points of respective clusters.
4. Repeat steps 2 and 3 until the cluster assignments do not change or a user-defined tolerance or a pre-defined maximum number of iterations is reached.

### Desnity-Based Spatial Clustering of Applications with Noise (DBSCAN)

DBSCAN is a density-based clustering algorithm that groups points when they are closely packed together (points with many nearby neighbors). Algorithm marks the data points as outliers when they lie alone in low-density regions (whose nearest neighbors are too far away).

1. Find the points in the $\mathcal{E}$ -neighborhood of every point, and identify the core points with more than minimum Data Points neighbors.
2. Find the connected components of core points on the neighbor graph, ignoring all non-core points.
3. Assign each non-core point to a nearby cluster if the cluster is an $\mathcal{E}$ -neighbor, otherwise assign it to noise.

### Hierarchical Clustering
Hierarchical clustering seeks to build a hierarchy of clusters.

**Agglomerative**, bottom-up approach where each observation starts in its own cluster, and pairs of clusters are merged as one moves up the hierarchy. Example draft algorithm:

1. Compute the distance matrix $D$ of all data points, i.e. $N\times N$ size.
2. Represent each data point as a singleton cluster.
3. Merge the two closest clusters based on the distance between the most dissimilar (distant) members of those two clusters.
4. Update the similarity matrix.
5. Repeat steps 2-4 until one single cluster remains

**Divisive**, top-down approach where all observations start in one cluster, and splits are performed recursively as one moves down the hierarchy. Example draft algorithm:
1. Assign all data points to one cluster.
2. For each cluster $k$ where $k=1,..,K$, find the most dissimilar (distant) member $x_{k1}$ to its centroid $\mu_k$  and find the most dissimilar member  $x_{k2}$ to $x_{k1}$
3. Identify the k with the highest dissimilarity between $x_k1$ and $x_k2$ where $k=1,..,K$
4. Make $x_{k1}$ and $x_{k2}$ new cluster centers, remove the original cluster center $\mu_k$ and re-assign all data points to the closest cluster centers.
5. Repeat steps 2-4 until each data point is a singleton.

Hierarchical methods cluster everything (either agglomerative or divisive) and let the user to decide about the number of clusters desired, since every possible number of clusters is already computed.

### Expectation Maximization
Expectation maximization algorithm is an iterative method to find maximum likelihood or maximum a posteriori (MAP) estimates of parameters (such as mean and standard deviation) in statistical models. Example draft algorithm:
1. Start with an initial estimate of what each parameter might be.
2. Compute the likelihood that each parameter produces the data point, i.e. compute the probability of each data point belonging to a particular cluster, which is a Gaussian distribution.
3. Calculate weights for each data point indicating whether it is more red or more blue based on the likelihood of it being produced by a parameter. Combine the weights with the data (expectation).
4. Compute a better estimate for the parameters using the weight-adjusted data (maximization).
5. Repeat steps 2-4 until the parameter estimate converges (the process stops producing a different estimate).

## Quality of Clustering - Metrics
### Cluster Separation

Inter-cluster distance = $E=\frac{1}{\left(\begin{array}{l}
K \\
2
\end{array}\right)} \sum_{k=1}^{K} \sum_{l=k+1}^{K} d_{k l}$
where $d_{kl}$ is the distance between cluster $k$ centroid and cluster $l$ centroid. Note that $d_{kk}$ is zero. A good clustering algorithm maximizes this metric.

### Cluster Cohesion
Intra-cluster distance= $E_{k}=\frac{1}{\left|x_{i} \in C_{k}\right|} \sum_{\forall x_{i} \in C_{k}}^{N} d_{i k}$, for cluster $C_k$

$E_k$ is the average distance between the cluster centroid and the data points of that cluster. A good clustering algorithm minimizes this metric.

### Scoring CLustering
Regular accuracy, i.e. Accuracy $=\frac{\# \text { correct }}{\# \text { total }}$

### Rand Index
After a clustering the cluster indexes may not match the original cluster indexes literally. The `contingency_matrix` function from the library `sklearn.metrics.cluster` can be used to map the resulting clusters to the original clusters (i.e. ground truth).

Given a set of $n$ elements $S=\left\{o_{1}, \ldots, o_{n}\right\}$ and two partitions of $S$ to compare, $X=\left\{X_{1}, \ldots, X_{r}\right\}$, a partition of $S$ into $r$ subsets, and $Y=\left\{Y_{1}, \ldots, Y_{s}\right\}$, a partition of $S$ into $s$ subsetss, define the following:

* $a$, the number of pairs of elements in $S$ that are in the same subset in $X$ and in the same subset in $Y$
* $b$, the number of pairs of elements in $S$ that are in different subsets in $X$ and in different subsets in $Y$
* $c$, the number of pairs of elements in $S$ that are in the same subset in $X$ and in different subsets in $Y$
* $d$, the number of pairs of elements in $S$ that are in different subsets in $X$ and in the same subset in $Y$

The Rand index, $R=\frac{a+b}{a+b+c+d}=\frac{a+b}{\left(\begin{array}{l}n \\ 2\end{array}\right)}$

Intuitively, $a+b$ can be considered as the number of **agreements** between $X$ and $Y$ and $c+d$ as the numebr of **disagreements** between $X$ and $Y$.

Since the denominator is the total number of pairs, the Rand index represents the frequency of occurrence of agreements over the total pairs, or the probability that $X$ and $Y$ will agree on a randomly chosen pair.

### Worked Example

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs

def get_minmax(_X, _m):  # _m = margin for visuals
    return _X[:,0].min()-_m, _X[:,0].max()+_m, _X[:,1].min()-_m, _X[:,1].max()+_m

def plotX(_X, _y, ax=None):  # Max 5 clusters/classes
    Colors = ['darkorchid', 'darkgreen', 'darkgrey', 'darksalmon', 'darkseagreen']
    Markers = ['o', '+', 'x', '1', '.']
    for c in range(max(_y)+1):
        plt.scatter(_X[_y==c,0], _X[_y==c,1], marker=Markers[c], color=Colors[c])
    #
    if ax is not None:
        ax.get_xaxis().set_ticks([])
        ax.get_yaxis().set_ticks([])
    #
    x1_min, x1_max, x2_min, x2_max = get_minmax(_X, 1)
    plt.xlim(x1_min, x1_max)
    plt.ylim(x2_min, x2_max)
    plt.xlabel(r'$x_1$', horizontalalignment='right', x=1.0)
    plt.ylabel(r'$x_2$', horizontalalignment='right', y=1.0)

# 5 classes with 2 features
X, y = make_blobs(n_samples=(100, 500, 500, 500, 1000), n_features=2, random_state=1,
                  cluster_std=[2, .5, .5, .5, 5],
                  centers=[(-3, -3), (-3, 3), (3, -3), (3, 3), (6, 6)])

plotX(X, y)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics.cluster import contingency_matrix
from sklearn.metrics import adjusted_rand_score

def kmeans_k(_n_clusters, _X, _y):
    km = KMeans(n_clusters=_n_clusters, init='random', n_init=10, max_iter=300, random_state=42)
    return km.fit_predict(_X), km.inertia_

def get_matchratio(_y, _y_km):
    # Compute contingency matrix
    cm = np.argmax(contingency_matrix(_y_km, _y), axis=1)
    # Map the clusters to the original using contingency above
    predicted = np.array([cm[c] for c in _y_km])
    # Compute match ratio, or accuracy
    match = sum([1 for y1, y2 in zip(_y, predicted) if y1 == y2])
    # Return the accuracy and mapped predictions
    return match/len(_y), predicted

EXPERIMENTS_K = (2,3,4,5,6,8,10)

for k in EXPERIMENTS_K:
    y_pred, inertia = kmeans_k(k, X, y)
    print(f'K-means with k={k:02d} performance, match rate={get_matchratio(y, y_pred)[0]:.3f}, '+
          f'Adjusted Rand index={adjusted_rand_score(y, y_pred):.3f}')
```

```python
K = 30
# Compute accuracies for demonstration of the elbow
acc, inertia = [], []
for k in range(1,K):
    y_pred, iner = kmeans_k(k, X, y)
    ac, pred = get_matchratio(y, y_pred)
    acc += [ac]
    inertia += [iner]
```

```python
# Accuracy
plt.plot(np.arange(1,K), acc, label='Accuracy')
plt.xlabel('k Number of Clusters')
k, ac = 9, 0.88
plt.annotate(f"elbow, k={k}", xy=(k, ac), xytext=(k+3, .9*ac), arrowprops=dict(facecolor='gold', shrink=1))
k, ac = 5, 0.77
plt.annotate(f"original, k={k}", xy=(k, ac), xytext=(k+3, .9*ac), arrowprops=dict(facecolor='gold', shrink=1))
plt.ylabel('Clustering Accuracy')
plt.legend(loc='lower left')

# Inertia
ax2 = plt.twinx()
ax2.plot(np.arange(1,K), inertia, color='firebrick', label='Inertia')
k, iner = 5, 24000
plt.annotate(f"inertia elbow, k={k}", xy=(k, iner), xytext=(k+3, 1.2*iner), arrowprops=dict(facecolor='coral', shrink=1))
plt.ylabel('Inertia')
plt.legend(loc='lower right')

plt.show()
```

**Important:** Notice that by using contingency mapping we are not reporting the true clustering error when we assign different cluster IDs to different parts of he same cluster. Thus, the following modification to the accuracy computation is realized, albeit probably penalizing the additional clusterings harshly.

Once a `kmeans` cluster is assigned to a particular original cluster, that original cluster ID is not used any more. All remaining unassigned clusters are assigned to new cluster IDs, so that error will accumulate.

```python
def get_matchratio2(_y, _y_km):
    # Compute contingency matrix
    cm = np.argmax(contingency_matrix(_y_km, _y), axis=1)
    # Modification to contingency cluster mapping
    k_new = len(np.unique(_y))  # assumes y has consequent labels of 0, 1, 2, ..
    gt_dict = {k:0 for k in range(k_new)}
    for i, c in enumerate(cm):
        if gt_dict[c] == 0:
            gt_dict[c] = 1
        else:
            cm[i] = k_new
            k_new += 1
    # Map the clusters to the original using contingency above
    predicted2 = np.array([cm[c] for c in _y_km])
    # Compute match ratio, or accuracy
    match = sum([1 for y1, y2 in zip(_y, predicted2) if y1 == y2])
    # Return the accuracy and mapped predictions
    return match/len(_y), predicted2
```

```python
# Cluster with k above
y_pred2 = get_matchratio2(y, kmeans_k(k, X, y)[0])[1]

# Plot resulting clustering
plt.scatter(X[:,0], X[:,1], c=y_pred2)
plt.show()
```

```python
for k in EXPERIMENTS_K:
    y_pred, inertia = kmeans_k(k, X, y)
    print(f'K-means with k={k:02d} performance, match rate={get_matchratio2(y, y_pred)[0]:.3f}, '+
          f'Adjusted Rand index={adjusted_rand_score(y, y_pred):.3f}')
```

```python
from collections import defaultdict
from functools import partial

def get_centroids(_X, _y):
    U = defaultdict(partial(np.ndarray, 0))
    Cn = defaultdict(int)
    for x, c in zip(_X, _y):
        U[c] = x if Cn[c] == 0 else U[c] + x
        Cn[c] += 1
    for c in U.keys():
        U[c] /= Cn[c]
    return U

def distance_inter_cluster(_U, _dist):
    # Compute inter-cluster errors
    K = np.max([int(c) for c in _U.keys()]) + 1
    if K < 2:
        return 0
    err, n = 0, 0
    for i in range(K):
        for j in range(i+1,K):
            err += _dist(_U[i], _U[j])
            n += 1
    return err / n

def distance_intra_cluster(_X, _y_km, _dist):
    # Compute intra-cluster errors
    U = get_centroids(_X, _y_km)
    En = defaultdict(partial(np.ndarray, 0))
    Cn = defaultdict(int)
    for x, c in zip(_X, _y_km):
        En[c] = _dist(x, U[c]) if Cn[c] == 0 else En[c] + _dist(x, U[c])
        Cn[c] += 1
    for c in U.keys():
        En[c] /= Cn[c]
    return En
```

### Dendograms for Hierarchical Clustering

The following cell demonstrates a simple dendogram for uniform distances for three linkages types: {`ward`, `complete`, `single`}. Full set of distances could be computed by the clustering program when clusters are joined, i.e. the distances between cluster centroids.

The x-axis are the individual data points, i.e. elements to be clustered. The y-axis levels are the clustering levels which is determined by the number of data points $N$.

```python
from scipy.cluster.hierarchy import dendrogram

def plot_dendogram_iris(_model, titletext='', **kwargs):
    # uniform distances
    distances = np.arange(_model.children_.shape[0])
    # number of observations contained in each clustering level
    no_of_observations = np.arange(2, _model.children_.shape[0]+2)
    # linkage matrix
    linkage_matrix = np.column_stack([_model.children_, distances, no_of_observations]).astype(float)
    # dendogram plotting
    dendrogram(linkage_matrix, **kwargs)
    plt.title(f'Iris dataset Dendrogram ({titletext})')
    plt.xlabel('sample index')
    
# Plot
plt.figure(figsize=(18, 6))
#
ax=plt.subplot(1, 3, 1)
model = AgglomerativeClustering(linkage='ward')
model.fit_predict(X)
plot_dendogram_iris(model, titletext='ward')
plt.ylabel('distance')
    
ax=plt.subplot(1, 3, 2)
model = AgglomerativeClustering(linkage='complete')
model.fit_predict(X)
plot_dendogram_iris(model, titletext='complete')

ax=plt.subplot(1, 3, 3)
model = AgglomerativeClustering(linkage='single')
model.fit_predict(X)
plot_dendogram_iris(model, titletext='single')

plt.tight_layout()
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs

def get_minmax(_X, _m):  # _m = margin for visuals
    return _X[:,0].min()-_m, _X[:,0].max()+_m, _X[:,1].min()-_m, _X[:,1].max()+_m

def plotX(_X, _y, ax=None):  # Max 5 clusters/classes
    Colors = ['darkorchid', 'darkgreen', 'darkgrey', 'darksalmon', 'darkseagreen']
    Markers = ['o', '+', 'x', '1', '.']
    for c in range(max(_y)+1):
        plt.scatter(_X[_y==c,0], _X[_y==c,1], marker=Markers[c], color=Colors[c])
    #
    if ax is not None:
        ax.get_xaxis().set_ticks([])
        ax.get_yaxis().set_ticks([])
    #
    x1_min, x1_max, x2_min, x2_max = get_minmax(_X, 1)
    plt.xlim(x1_min, x1_max)
    plt.ylim(x2_min, x2_max)
    plt.xlabel(r'$x_1$', horizontalalignment='right', x=1.0)
    plt.ylabel(r'$x_2$', horizontalalignment='right', y=1.0)

# 5 classes with 2 features
X, y = make_blobs(n_samples=(100, 500, 500, 500, 1000), n_features=2, random_state=1,
                  cluster_std=[2, .5, .5, .5, 0.5],
                  centers=[(-3, -3), (-3, 3), (3, -3), (3, 3), (6, 6)])
plotX(X, y)
```

The 5th class is now centered on 1,1
